```python
# --- INSTALL DEPENDENCIES ---
!pip install torch torchvision torchaudio -q
!pip install torch-geometric -q
!pip install torch-geometric-temporal -q
!pip install matplotlib networkx seaborn scikit-learn -q

# --- IMPORTS ---
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric_temporal.dataset import ChickenpoxDatasetLoader
from torch_geometric.nn import GCNConv
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

# --- DATA LOADING ---
loader = ChickenpoxDatasetLoader()
dataset = list(loader.get_dataset())
snapshot = dataset[0]
input_dim = snapshot.x.shape[1]

# --- VISUALIZE GRAPH STRUCTURE ---
edge_index = torch.tensor(snapshot.edge_index, dtype=torch.long)
G = nx.DiGraph()
for i in range(edge_index.shape[1]):
    src, dst = edge_index[0, i].item(), edge_index[1, i].item()
    G.add_edge(src, dst)
plt.figure(figsize=(10, 6))
nx.draw_spring(G, node_size=80, arrows=True, alpha=0.7)
plt.title("Chickenpox Counties Graph")
plt.show()

# --- LTC LAYER ---
class LTCNeuronLayer(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.w = nn.Parameter(torch.randn(in_features, out_features))
        self.bias = nn.Parameter(torch.zeros(out_features))
        self.time_constant = nn.Parameter(torch.rand(out_features))

    def forward(self, x, t):
        delta = x @ self.w + self.bias
        tau = F.softplus(self.time_constant) + 1e-3
        membrane_state = delta * (1 - torch.exp(-1 * t / tau))
        return membrane_state

# --- LIQUID TOPOLOGY NETWORK ---
class LiquidTopoNetLTC(nn.Module):
    def __init__(self, in_dim, hidden_dim, out_dim):
        super().__init__()
        self.gcn = GCNConv(in_dim, hidden_dim)
        self.ltc = LTCNeuronLayer(hidden_dim, hidden_dim)
        self.out = nn.Linear(hidden_dim, out_dim)

    def forward(self, x, edge_index, t):
        x = F.relu(self.gcn(x, edge_index))
        x = self.ltc(x, t)
        return self.out(x)

# --- GCN BASELINE ---
class StandardGCN(nn.Module):
    def __init__(self, in_dim, hidden_dim, out_dim):
        super().__init__()
        self.conv1 = GCNConv(in_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, out_dim)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        return self.conv2(x, edge_index)

# --- TRAINING FUNCTION ---
def train(model, dataset, use_time=False):
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
    loss_fn = nn.MSELoss()
    losses = []
    for epoch in range(10):
        total_loss = 0
        for time, snapshot in enumerate(dataset):
            x = torch.tensor(snapshot.x, dtype=torch.float)
            edge_index = torch.tensor(snapshot.edge_index, dtype=torch.long)
            y = torch.tensor(snapshot.y, dtype=torch.float).unsqueeze(1)
            t = torch.tensor(float(time))

            optimizer.zero_grad()
            out = model(x, edge_index, t) if use_time else model(x, edge_index)
            loss = loss_fn(out, y)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        avg = total_loss / len(dataset)
        losses.append(avg)
        print(f"Epoch {epoch+1} | Loss: {avg:.4f}")
    return losses

# --- TRAIN MODELS ---
model_ltc = LiquidTopoNetLTC(input_dim, 32, 1)
losses_ltc = train(model_ltc, dataset, use_time=True)

model_gcn = StandardGCN(input_dim, 32, 1)
losses_gcn = train(model_gcn, dataset, use_time=False)

# --- COMPARE LOSS CURVES ---
plt.figure(figsize=(8, 4))
plt.plot(losses_ltc, label='LiquidTopoNet (LTC)', marker='o')
plt.plot(losses_gcn, label='Standard GCN', marker='x')
plt.title("Training Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.show()

# --- EVALUATION FUNCTION ---
def evaluate(model, dataset, use_time=False):
    model.eval()
    all_preds, all_true = [], []
    for time, snapshot in enumerate(dataset):
        x = torch.tensor(snapshot.x, dtype=torch.float)
        edge_index = torch.tensor(snapshot.edge_index, dtype=torch.long)
        y = torch.tensor(snapshot.y, dtype=torch.float).unsqueeze(1)
        t = torch.tensor(float(time))
        with torch.no_grad():
            pred = model(x, edge_index, t) if use_time else model(x, edge_index)
        all_preds.append(pred.squeeze().numpy())
        all_true.append(y.squeeze().numpy())
    return np.array(all_preds), np.array(all_true)

# --- GET PREDICTIONS ---
ltc_preds, true_vals = evaluate(model_ltc, dataset, use_time=True)
gcn_preds, _ = evaluate(model_gcn, dataset, use_time=False)

# --- PRINT METRICS ---
print(f"LTC Model → MSE: {mean_squared_error(true_vals, ltc_preds):.4f}, R²: {r2_score(true_vals, ltc_preds):.4f}")
print(f"GCN Model → MSE: {mean_squared_error(true_vals, gcn_preds):.4f}, R²: {r2_score(true_vals, gcn_preds):.4f}")

# --- VISUALIZATION: CORRELATION MATRIX ---
corr_df = pd.DataFrame({
    'True': true_vals.flatten(),
    'LTC': ltc_preds.flatten(),
    'GCN': gcn_preds.flatten()
})
plt.figure(figsize=(6, 4))
sns.heatmap(corr_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Prediction Correlation Matrix")
plt.show()

# --- VISUALIZATION: PREDICTION SCATTER ---
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.scatter(true_vals, ltc_preds, alpha=0.5, color='blue')
plt.xlabel("True")
plt.ylabel("LTC Predicted")
plt.title("LTC Predictions")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(true_vals, gcn_preds, alpha=0.5, color='green')
plt.xlabel("True")
plt.ylabel("GCN Predicted")
plt.title("GCN Predictions")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- VISUALIZATION: PER-NODE ERROR BARS ---
ltc_errors = ((ltc_preds - true_vals) ** 2).mean(axis=0)
gcn_errors = ((gcn_preds - true_vals) ** 2).mean(axis=0)

plt.figure(figsize=(12, 5))
x = np.arange(len(ltc_errors))
plt.bar(x - 0.2, ltc_errors, width=0.4, label='LTC', color='blue')
plt.bar(x + 0.2, gcn_errors, width=0.4, label='GCN', color='green')
plt.xlabel("Node Index")
plt.ylabel("Avg MSE")
plt.title("Per-Node Average Prediction Error")
plt.legend()
plt.grid(True)
plt.show()
```
